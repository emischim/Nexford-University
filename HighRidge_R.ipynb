```r
# First Step: I am creating a list of 400 workers dynamically
set.seed(123) #I am Setting seed to ensure that the generation is sequential
high_ridge_workers <- list()
for (i in 1:450) {  #I am making sure I generate at least 400 workers, so I randomly choose 450
    high_ridge_worker <- list(
        name = paste("High_Ridge_Worker", i, sep="_"), #Since the company is Highridge, I am appending numbers to "High_Ridge_Worker" as a means of identifying each worker
        salary = sample(7500:35000, 1),  # I am generating figures between the lowest salary of 7500 and highest salary of 35000
        gender = sample(c("male", "female"), 1) #I am generating random male and female selections , since it is required in the logic
    )
    high_ridge_workers[[i]] <- high_ridge_worker
}

# Third Step: Here, I am generating payment slips with conditions
for (i in 1:length(high_ridge_workers)) {
    tryCatch({
        high_ridge_worker <- high_ridge_workers[[i]]

        if (high_ridge_worker$salary > 10000 & high_ridge_worker$salary < 20000) {
            high_ridge_worker$level <- "A1"
        }
        if (high_ridge_worker$salary > 7500 & high_ridge_worker$salary < 30000 & high_ridge_worker$gender == "female") {
            high_ridge_worker$level <- "A5-F"
        } #In this block of code, I am assigning levels based on salary and gender as described in the assignment which are:
          #If the salary is greater than $10,000 and less than $20,000, I assign the Employee level as "A1."
          #If the salary is greater than $7,500 and less than $30,000 and the employee is female, I set the Employee level as "A5-F."

        payment_slips <- list(
            name = high_ridge_worker$name,
            salary = high_ridge_worker$salary,
            level = ifelse(is.null(high_ridge_worker$level), "No level assigned", high_ridge_worker$level) # In this block of code, I am generating the payment slips
        )
        print(payment_slips)   # Printing the payment slips

    }, error = function(e) {
        cat("Error processing worker", high_ridge_worker$name, ":", e$message, "\n") # I am addressing potential errors with exception handling
    })
}
```
