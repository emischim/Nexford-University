```python
#First Step :I am importing the random module ot be able to generate the list of workers dynamically
import random

#Second Step: Here, I would create a list of 400 workers dynamically.
high_ridge_workers = []
for i in range(450): #I am making sure I generate at least 400 workers, so I randomly choose 450
    high_ridge_worker = {
        "name": f"High_Ridge_Worker_{i+1}", #Since the company is Highridge, I am appending numbers to "High_Ridge_Worker" as a means of identifying each worker
        "salary": random.randint(7500, 35000), # I am generating figures between the lowest salary of 7500 and highest salary of 35000
        "gender": random.choice(["male", "female"]) #I am generating random male and female selections , since it is required in the logic
    }
    high_ridge_workers.append(high_ridge_worker)

# Third Step: Here, I am generating payment slips with conditions
for high_ridge_worker in high_ridge_workers:
    try:
        if 10000 < high_ridge_worker["salary"] < 20000:
            high_ridge_worker["level"] = "A1"
        if 7500 < high_ridge_worker["salary"] < 30000 and high_ridge_worker["gender"] == "female":
            high_ridge_worker["level"] = "A5-F" #In this block of code, I am assigning levels based on salary and gender as described in the assignment which are:
                                                #If the salary is greater than $10,000 and less than $20,000, I assign the Employee level as "A1."
                                                #If the salary is greater than $7,500 and less than $30,000 and the employee is female, I set the Employee level as "A5-F."


        payment_slips = {
            "name": high_ridge_worker["name"],
            "salary": high_ridge_worker["salary"],
            "level": high_ridge_worker.get("level", "No level assigned")  # In this block of code, I am generating the payment slips
        }
        print(payment_slips)  # Printing the payment slips

    except Exception as e:
        print(f"Error processing worker {high_ridge_worker['name']}: {e}")  # I am addressing potential errors with exception handling
```
